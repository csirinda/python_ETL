# 3.1 Renaming Columns

```python
#ขียน loop เพื่อสร้าง list ของชื่อคอลัมน์ใหม่ที่ clean แล้ว
#ใส่ตัวแปร new_column_names
new_column_names = []
import re
for col_name in df.columns:
        new_col_name = re.sub('\W+','_', col_name) #แทนที่สิ่งที่ไม่ใช่ตัวอักษรหรือตัวเลขด้วย _
        if not new_col_name[-1].isalnum(): #ถ้าตัวอักษรตัวสุดท้ายไม่ใช่ตัวอักษรหรือตัวเลข
            new_col_name = new_col_name[:-1]
        new_col_name = new_col_name.lower() #เปลี่ยนเป็นตัวพิมพ์เล็ก
        new_column_names.append(new_col_name) #ยัดใส่ list ที่สร้างไว้
```

```python
#เช็คว่าชื่อคอลัมน์เป็นไปตามที่ต้องการหรือยัง
print(new_column_names)
```

```python
#เปลี่ยนชื่อคอลัมน์ตาม new_column_names
df.columns = new_column_names
```

# 3.2 Changing Format

เริ่มจากดูว่า column year มี format แบบไหนบ้าง
จากนั้นเปลี่ยนให้เป็นแบบเดียวกันให้หมด

```python
df['year'].unique()
```

```python
#เอาแค่เลขปี ค.ศ.
df['year'] = df.year.apply(lambda x: str(str(x).split("(")[-1]).split(")")[0])
years= df.year

## years that end in "-" means that the show is still going,so its better to take this account in order to update it later on to current year.
with_hypen=[]
for i in range(len(years)):
    if years[i][-1]==" ":
        with_hypen.append(1)
    else:
        with_hypen.append(0)
df['hypen_year']=with_hypen

## Getting only the interger values and joining the integere values to create year-like pattern YYYY-YYYY and separating them into differetn columns
def convert(list):

    # Converting integer list to string list
    s = [str(i) for i in list]

    # Join list items using join()
    res = "".join(s)

    return(res)

year=[]
year_1=[]
year_2=[]
for i in range(len(df)):
    a = re.findall("\w\d",df.year[i])
    year.append(convert(a))
    year_1.append(year[i][:4])
    year_2.append(year[i][-4:])
df['year']=year
df['beg_year']=year_1
df['end_year']=year_2

## Converting values to numeric values
df['beg_year']=pd.to_numeric(df['beg_year'])
df['end_year']=pd.to_numeric(df['end_year'])

# updating all end years with - (eg:-2021-) to 2022
hypen=[]
for i in range(len(df)):
    if df['hypen_year'][i]==1:
        hypen.append(2022)
    else:
        hypen.append(df['end_year'][i])
df['end_year']=hypen

df['Num_years']=df['end_year']-df['beg_year']
df.beg_year.fillna(0,inplace=True)
df.end_year.fillna(0,inplace=True)
df.Num_years.fillna(0,inplace=True)
```

```python
# เรียกมาดู
df['year'].unique()
```

ลบ \n ในทุก colume ที่มี

```python
# ตัด \n ออกให้หมด
for col in ['genre','one_line','stars']:
    df[col] = df[col].str.replace("\n","").str.strip()

df.head()
```

สร้าง column ใหม่ จากstars แยกระหว่างdirectors กับstars

```python
df['director']=df['stars'].apply(lambda x: str(x).split("Stars")[0]) #สร้าง col director โดยแยกหน้า-หลัง Stars เก็บค่าด้านหน้าไว้
df['actors'] = df['stars'].apply(lambda x: str(x).split("Stars")[-1]) #สร้าง col actors โดยแยกหน้า-หลัง Stars เก็บค่าด้านหลังไว้

df['actors']=df['actors'].str.replace(":","")
df['director']= df['director'].apply(lambda x: x.split(":")[-1]).str.replace("|","")
```

# 3.3 Dealing with Missing Values

เมื่อเราเจอข้อมูลที่เป็นค่าว่าง (เช่น NaN = Not a Number, Null, NA = Not Available) เราสามารถที่จะ

* เติมค่า (fill) เมื่อมีค่าที่ถูกต้องหรือเหมาะสมให้เติม
แทนค่า (impute) เมื่อต้องใส่ "ค่าตัวแทน" เพื่อให้เข้าโมเดล machine learning ได้
* เอาออก (drop) เมื่อข้อมูลหายเป็นจำนวนมาก แล้วที่เหลืออยู่ก็ใช้อะไรไม่ค่อยได้
* ปล่อยไว้ (ignore) เมื่อไม่มีค่าที่เหมาะสมให้เติม แล้วเราก็ไม่ได้ต้องเอาข้อมูลไปเข้าโมเดลอะไร

เติมค่า Median แทนค่า NaN ใน Rating, Votes, Runime

```python
df['votes']=df['votes'].str.replace(",","")
df['votes']=pd.to_numeric(df['votes'])
df['votes']=df['votes'].fillna(df.votes.median())

df['runtime']=df['runtime'].fillna(df.runtime.median())

df['rating']=df['rating'].fillna(df.rating.median())

df['gross'].fillna("0",inplace=True)
```
